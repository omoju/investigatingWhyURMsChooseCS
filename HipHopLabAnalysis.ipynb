```python
from __future__ import division


def scaleData(theDataFrame):
    # must sort to get index in order
    # I used the sum of everything so that we are comparing real percentages.
    theMin, theMax = theDataFrame.min(), theDataFrame.sum(axis=1) 
 
    for (index, val) in theDataFrame.iteritems():
        theDataFrame[index] = ( val / theMax) * 100
        
    #Rounding errors doesn't make everything add up to 100, sometimes its 98, 97 and so on
    
def turnArray(x):
    theList = []
    if 1 in x.index:
        ("")
    else:
        x.set_value(1, 0)

    if 2 in x.index:
        ("")
    else:
        x.set_value(2, 0)

    if 3 in x.index:
        ("")
    else:
        x.set_value(3, 0)

    if 4 in x.index:
        ("")
    else:
        x.set_value(4, 0)

    if 5 in x.index:
        ("")
    else:
        x.set_value(5, 0)
        
    for i in range(1, len(x)+1):
        theList.append(x[i])
    a = array( theList ) 
    return a


# Customizing the figure
font = {'family' : 'serif',
        'color'  : '#141433',
        'weight' : 'normal',
        'size'   : 12,
        }

params = {
   'axes.labelsize': 8,
   'text.fontsize': 8,
   'legend.fontsize': 10,
   'xtick.labelsize': 10,
   'ytick.labelsize': 10,
   'text.usetex': False,
   'figure.figsize': [5.5, 4.5]
   }
    

plot_params = {
    'xmin_gen' : 0,
    'xmax_gen' : 0,
    'ymin_gen' : 0,
    'ymax_gen' : 0,
    'xlabel' : '',
    'ylabel' : '',
    'title' : '',
    'legend_label': ['',''],
    'x1Color' : 'r',
    'x2Color' : 'b'
}

#Get the description of the column items
dd = pd.read_csv('Data/Data_Describe.csv')
dd.columns = ['dataDecription', 'dataKeys']
dataDescription = {}

for i, row in dd.iterrows():
    #print i, row.dataKeys, row.dataDecription
    dataDescription[dd.dataKeys[i]] = dd.dataDecription[i]
    
print '\n'
del dataDescription['name_2']
del dataDescription['name_1']
del dataDescription['major']
del dataDescription['morecs']
del dataDescription['timestamp']
del dataDescription['consent']
del dataDescription['mtr_2']
del dataDescription['mtr_3']
del dataDescription['mtr_1']
del dataDescription['prcs_3']
del dataDescription['prcs_5']
del dataDescription['prcs_4']
del dataDescription['prcs_2']
del dataDescription['prcs_1']
del dataDescription['reason_class']
del dataDescription['grade']
del dataDescription['gender']
del dataDescription['prepared']

#'snap_python','hiphop_d1','hiphop_d2','song_ct'


dataDescription['atct_1'] = 'I am Good at Solving a Problem by Thinking \nabout Similar Problems I have Solved Before.'
dataDescription['atct_7'] = 'I am Good at Ignoring Irrelevant Details to Solve a Problem.'
dataDescription['cltrcmp_2'] = 'I have Good Cultural Competence, or the Ability to Interact Effectively \n with People from Diverse Backgrounds.'


dataDescription['snap_python'] = 'I was able to see how computational thinking \ncarries over from Snap! to Python.'
dataDescription['hiphop_d1'] = 'The hip-hop data module made me see how computation \ncan enhance culture.'
dataDescription['hiphop_d2'] = 'Lyrical analysis gave me a new way to think about computation.'
dataDescription['song_ct'] = 'I was able to build on my pre-existing knowledge about songs \nto further understand computation.'
#for key in dataDescription:
#    print key, dataDescription[key]


def plot_data(ax, x1, plot_params):

    ind = np.arange(plot_params['xmax_gen'])  # the x locations for the groups
    width = 0.5       # the width of the bars

    rects1 = ax.bar(ind, x1, width, color = plot_params['x1Color'])
    
    # change xlim to set_xlim
    ax.set_xlim(plot_params['xmin_gen'], plot_params['xmax_gen'])
    ax.set_ylim(plot_params['ymin_gen'], plot_params['ymax_gen'])

    #change xticks to set_xticks
    ax.set_xticks(ind+0.3)
    ax.set_xticklabels( ('Not Really', '', 'Neutral', '', 'Absolutely') )
    ax.set_yticks(np.arange(plot_params['ymin_gen'], plot_params['ymax_gen']+10, 10))
    legend = ax.legend(plot_params['legend_label'], loc=9);
    ax.set_xlabel(plot_params['xlabel'], fontdict=font)
    ax.set_ylabel(plot_params['ylabel'], fontdict=font)
    ax.set_title(plot_params['title'], fontdict=font)
    ax.legend(loc=9)

    def autolabel(rects):
    # attach some text labels
        for rect in rects1:
            height = rect.get_height()
            ax.text(rect.get_x()+rect.get_width()/2., 1.05*height, '%d'%int(height), ha='center', va='bottom')

    autolabel(rects1)
    
    # Pad margins so that markers don't get clipped by the axes
    ax.margins(0.1)
   
```

```python
def plot_data2(ax, x1, x2, plot_params):

    ind = np.arange(plot_params['xmax_gen'])  # the x locations for the groups
    width = 0.35       # the width of the bars

    rects1 = ax.bar(ind, x1, width, color = plot_params['x1Color'])
    rects2 = ax.bar(ind+width, x2, width, color = plot_params['x2Color'])

    # change xlim to set_xlim
    ax.set_xlim(plot_params['xmin_gen'], plot_params['xmax_gen'])
    ax.set_ylim(plot_params['ymin_gen'], plot_params['ymax_gen'])

    #change xticks to set_xticks
    ax.set_xticks(ind+width)
    ax.set_xticklabels( ('Not Really', '', 'Neutral', '', 'Absolutely') )
    ax.set_yticks(np.arange(plot_params['ymin_gen'], plot_params['ymax_gen']+10, 10))
    legend = ax.legend(plot_params['legend_label'], loc=9);
    ax.set_xlabel(plot_params['xlabel'], fontdict=font)
    ax.set_ylabel(plot_params['ylabel'], fontdict=font)
    ax.set_title(plot_params['title'], fontdict=font)
    ax.legend(loc=9)

    def autolabel(rects):
    # attach some text labels
        for rect in rects:
            height = rect.get_height()
            ax.text(rect.get_x()+rect.get_width()/2., 1.05*height, '%d%%'%int(height), ha='center', va='bottom')

    autolabel(rects1)
    autolabel(rects2)
    
    # Pad margins so that markers don't get clipped by the axes
    ax.margins(0.1)
   
```

```python
def genFigure(itemDimension, theCategory, theXLabel):
    
    temp =[]
    for key in itemDimension:
        temp.append('cs10'+theCategory+'.'+key+'.value_counts()')
        a = eval(temp[0])
        scaleData(a)
        x0 = turnArray(a)
        temp = []
        
        
        #update params figure size to accomodate 2 figures
        params['figure.figsize'] = [6.5, 4.5]

        plt_0_params = {
            'xmin_gen' : 0.0,
            'xmax_gen' : 5.0,
            'ymin_gen' : 0,
            'ymax_gen' : 100,
            'xlabel' : theXLabel[0],
            'ylabel' : 'Distribution of Respondents',
            'title' : '',
            'legend_label': ["CS10"],
            'x1Color' : colors_1[1],
            'x2Color' : colors_1[0]
            }

       


        rcParams.update(params)
        fig = figure(dpi=80) # no frame
        ax0 = fig.add_subplot(111)
        fig.suptitle(dataDescription[key], fontdict=font)

        plot_data(ax0, x0, plt_0_params)
        
        fileName = []
        fileName.append(key+'_CS10'+theCategory+'.pdf')
        fig.savefig(fileName[0], dpi=72)
        rcParams.update(params)
```

```python
def genderMaleFemale(itemDimension, theCategory, theXLabel):
    
    temp =[]
    for key in itemDimension:
        temp.append('cs10'+theCategory+'_female'+'.'+key+'.value_counts()')
        b = eval(temp[0])
        scaleData(b)
        x0_female = turnArray(b)
        temp = []
        
        temp.append('cs10'+theCategory+'_male'+'.'+key+'.value_counts()')
        b = eval(temp[0])
        scaleData(b)
        x0_male = turnArray(b)
        temp = []
        
        #update params figure size to accomodate 2 figures
        params['figure.figsize'] = [6.5, 4.5]
        plt_0_params = {
            'xmin_gen' : 0.0,
            'xmax_gen' : 5.0,
            'ymin_gen' : 0,
            'ymax_gen' : 100,
            'xlabel' : theXLabel[0],
            'ylabel' : 'Distribution of Respondents',
            'title' : '',
            'legend_label': ['CS10 Female', 'CS10 Male'],
            'x1Color' : colors_1[2],
            'x2Color' : colors_2[9]
        }


        rcParams.update(params)
        fig = figure(dpi=80) # no frame
        ax0 = fig.add_subplot(111)
        #print key, dataDescription[key]
        fig.suptitle(dataDescription[key], fontdict=font)
        
        plot_data2(ax0, x0_female, x0_male, plt_0_params)
        
        fileName = []
        fileName.append(key+'_malefemale'+theCategory+'.pdf')
        fig.savefig(fileName[0], dpi=72)
        rcParams.update(params)

       
    



def genderMaleFigure(itemDimension, theCategory, theXLabel):
    
    temp =[]
    for key in itemDimension:
        temp.append('cs10'+theCategory+'_male'+'.'+key+'.value_counts()')
        b = eval(temp[0])
        scaleData(b)
        x0_male = turnArray(b)
        temp = []
  
        #update params figure size to accomodate 2 figures
        params['figure.figsize'] = [6.5, 4.5]
        plt_0_params = {
            'xmin_gen' : 0.0,
            'xmax_gen' : 5.0,
            'ymin_gen' : 0,
            'ymax_gen' : 100,
            'xlabel' : theXLabel[1],
            'ylabel' : 'Distribution of Respondents',
            'title' : '',
            'legend_label': ['CS10'],
            'x1Color' : colors_2[4],
            'x2Color' : colors_2[9]
        }


        rcParams.update(params)
        fig = figure(dpi=80) # no frame
        ax0 = fig.add_subplot(111)
        #print key, dataDescription[key]
        fig.suptitle(dataDescription[key], fontdict=font)
        
        plot_data(ax0, x0_male, plt_0_params)
        
        
        fileName = []
        fileName.append(key+'_male'+theCategory+'.pdf')
        fig.savefig(fileName[0], dpi=72)
        rcParams.update(params)

```
